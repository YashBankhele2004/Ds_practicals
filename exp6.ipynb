```python
import pandas as pd
import matplotlib
df= pd.read_csv ("/content/cats_dataset.csv")
df
```

```python
import matplotlib.pyplot as plt


plt.hist(df['Age (Years)'], bins=range(0, 21, 1), edgecolor='black')
plt.xlabel('Age (Years)')
plt.ylabel('Frequency')
plt.title('Distribution of Cat Ages')
plt.show()
```

```python
plt.scatter(df['Age (Years)'], df['Weight (kg)'])
plt.xlabel('Age (Years)')
plt.ylabel('Weight (kg)')
plt.title('Weight vs. Age of Cats')
plt.show()
```

```python
avg_weights = df.groupby('Breed')['Weight (kg)'].mean()
avg_weights.plot(kind='bar')
plt.xlabel('Breed')
plt.ylabel('Average Weight (kg)')
plt.title('Average Weight by Cat Breed')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_weight_by_age = df.groupby('Age (Years)')['Weight (kg)'].mean()

plt.plot(avg_weight_by_age.index, avg_weight_by_age.values, marker='o', linestyle='-', color='b')
plt.xlabel('Age (Years)')
plt.ylabel(' Weight (kg)')
plt.title('Weight by Age')
plt.grid(True)
plt.show()
```

```python
bins = range(0, df['Age (Years)'].max() + 1, 2)
df['Age Group'] = pd.cut(df['Age (Years)'], bins=bins)

plt.figure(figsize=(12, 6))
df.boxplot(column='Weight (kg)', by='Age Group', grid=False)
plt.xlabel('Age Group')
plt.ylabel('Weight (kg)')
plt.title('Weight Distribution by Age Group')
plt.suptitle('')
plt.xticks(rotation=45)
plt.show()
```
