```python
features = df_apple.columns.tolist()
plt.figure(figsize=(20, 15))
for i, feature in enumerate(features):
    plt.subplot(6, 4, i + 1)
    sns.boxplot(data=df_apple[feature])
    plt.title(f'Box plot of {feature}')
plt.tight_layout()
plt.show()
```

```python
numeric_columns = df_apple.select_dtypes(include=['float64', 'int64']).columns
Q1 = df_apple[numeric_columns].quantile(0.25)
Q3 = df_apple[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
df = df_apple[~((df_apple[numeric_columns] < (Q1 - 1.5 * IQR)) | (df_apple[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
df
```
