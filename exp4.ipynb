```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
file_path = '/IEA Global EV Data 2024.csv'
df = pd.read_csv(file_path)
print("Original Data:")
print(df.head())
print("\nData Types:")
print(df.dtypes)
print("\nSummary Statistics:")
print(df.describe())

numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Min-Max Normalization
min_max_scaler = MinMaxScaler()
normalized_data = min_max_scaler.fit_transform(df[numeric_columns])

# Convert the normalized data back to a DataFrame
normalized_df = pd.DataFrame(normalized_data, columns=numeric_columns)

# Replace the original numeric columns with the normalized ones
df[numeric_columns] = normalized_df

print("\nNormalized Data (Min-Max Scaling):")
print(df.head())

# Standardization (Z-score Normalization)
standard_scaler = StandardScaler()
standardized_data = standard_scaler.fit_transform(df[numeric_columns])

# Convert the standardized data back to a DataFrame
standardized_df = pd.DataFrame(standardized_data, columns=numeric_columns)

# Replace the original numeric columns with the standardized ones
df[numeric_columns] = standardized_df
print("\nStandardized Data (Z-score Normalization):")
print(df.head())
```
