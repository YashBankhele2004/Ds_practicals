```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.utils import resample
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/apple_quality.csv")
df
```

```python
df['Size'] = pd.to_datetime(df['Size'])

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Size'], df['Weight'], marker='o', linestyle='-', color='b')

# Adding title and labels
plt.title('Apple Size over Weight')
plt.xlabel('Size')
plt.ylabel('Weight')

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Display the line chart
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Plot each attribute
plt.plot(df.index, df['Size'], marker='o', linestyle='-', label='Size')
plt.plot(df.index, df['Ripeness'], marker='o', linestyle='-', label='Ripeness')
plt.plot(df.index, df['Weight'], marker='o', linestyle='-', label='Weight')
plt.plot(df.index, df['Sweetness'], marker='o', linestyle='-', label='Sweetness')
plt.plot(df.index, df['Crunchiness'], marker='o', linestyle='-', label='Crunchiness')

# Adding title and axis labels
plt.title('Apple Quality Attributes')
plt.xlabel('Observation Index')  # X-axis label
plt.ylabel('Attribute Value')     # Y-axis label

# Adding a legend to the plot
plt.legend()

# Display the grid and line chart
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the line for the 'Size' attribute
line, = ax.plot(df.index, df['Size'], label='Size')

# Place a marker at a specific index (e.g., index 10)
marker_index = 10
marker, = ax.plot(df.index[marker_index], df['Size'].iloc[marker_index], 'o', color=line.get_color(), markersize=10)

# Adding title and axis labels
ax.set_title('Apple Size Attribute with Marker')
ax.set_xlabel('Index')
ax.set_ylabel('Size')

# Show the legend
ax.legend()

# Display the plot
plt.show()
```

```python
import seaborn as sns

# Create bins for a categorical version of the 'Sweetness' attribute
bins = [0, 2, 4, 6, 8, 10]  # Example bins for sweetness
labels = ['Low', 'Moderate', 'High', 'Very High', 'Extremely High']
df['Sweetness_Category'] = pd.cut(df['Sweetness'], bins=bins, labels=labels, right=False)

# Create a count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Sweetness_Category', data=df)

# Adding title and axis labels
plt.title('Count of Apple Sweetness Categories')
plt.xlabel('Sweetness Category')
plt.ylabel('Count')

# Display the plot
plt.show()
```
